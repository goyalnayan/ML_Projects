```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
X = df.iloc[:, 0:3].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Residual
ypred_test = model.predict(X_test)
ypred_test
```

```python
df.head(2)
```

## 1. Linear Relationship

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(13, 9))
ax1.scatter(df['TV'], df['sales'])
ax1.set_title('sales1')
ax2.scatter(df['radio'], df['sales'])
ax2.set_title('sales1')
ax3.scatter(df['newspaper'], df['sales'])
ax3.set_title('sales1')
```

```python
plt.figure(figsize=(5, 4))
residual = y_test - ypred_test
plt.scatter(y_test, residual)
plt.xlabel('observed_values')
plt.ylabel('fitted_values')
plt.show()
```

## 2. Multicollinearity (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = []
for i in range(X_train.shape[1]):
    vif.append(variance_inflation_factor(X_train, i))

pd.DataFrame({'vif':vif}, index=df.columns[0:3]).T
```

## 3. Normality of Errors

```python
residual = y_test - ypred_test
```

```python
sns.displot(residual, kde='True')
```

```python
# QQ plot
import scipy as sp
fig, ax = plt.subplots(figsize=(6, 4))
sp.stats.probplot(residual, plot=ax, fit=True)
plt.show()
```

## 4. Homoscadescity

```python
sns.boxplot(df['newspaper'])
```

```python
# Calculate Q1
Q1 = df['newspaper'].quantile(0.25)
print("Q1: ", Q1)

# Calculate Q2
Q3 = df['newspaper'].quantile(0.75)
print("Q3: ", Q3)

# Calculate IQR
IQR = Q3 - Q1
print(IQR)

# Calculate lower limit of outliers
lower_limit = Q1 - (IQR * 1.5)
print(lower_limit)

# Calculate upper limit of outliers
upper_limit = Q3 + (IQR * 1.5)
print(upper_limit)
```

```python
df[(df['newspaper'] < lower_limit) | (df['newspaper'] > upper_limit)]
```

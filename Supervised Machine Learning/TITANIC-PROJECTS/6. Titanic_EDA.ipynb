```python
df['family_size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.head(2)
```

```python
# family_type
def cal(number):
    if number == 1:
        return 'alone'
    elif number > 1 and number < 5:
        return 'small'
    else:
        return 'large'
```

```python
df['family_type'] = df['family_size'].apply(cal)
df
```

now, you have created a new column so now you can perform univaraite or bivaraite analysis on this column

```python
pd.crosstab(df['Survived'], df['family_type'], normalize='columns') * 100
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = [["milk",'bread','rice','book'],['bread','jam','book','pen'],
        ['jam','milk','bread','rice','egg'],['rice','egg','pen','book'],
        ['egg','pen','milk','bread','jam'],['egg','rice','bread','jam']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_array = te.fit_transform(data)
```

```python
import pandas as pd
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
itemset = apriori(df, min_support=0.5, use_colnames=True)
itemset
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(itemset, metric='confidence', min_threshold=0.6)
res
```

```python
result = res[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
result
```
